```
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt


import sklearn 
from sklearn import preprocessing
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import confusion_matrix
```

```
df = pd.read_csv('/content/cc_approvals.data',header=None)
df.head()
```

```
df.isnull().sum()
```

```
df.replace(to_replace='?',value = np.NaN,inplace=True)
```

```
df.isnull().sum()
```

```
for col in df.columns:
  if df[col].dtypes == 'object':
    df =df.fillna(df[col].value_counts().index[0])
df.isnull().sum()
```

```
enc = preprocessing.LabelEncoder()
for col in df.columns:
  if df[col].dtypes == 'object':
    df[col]=enc.fit_transform(df[col])
df.head()
```

```
df.drop([11,13],inplace=True,axis=1)
df=df.values
df.shape
x =df[:,0:13]
y=df[:,13]

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state=42)
```

```
from sklearn.metrics import accuracy_score
scaler = preprocessing.MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
lre = LogisticRegression(C=0.1,solver ='liblinear')
lre.fit(x_train,y_train)
y_pred=lre.predict(x_test)
lre.score(x_test,y_test),accuracy_score(y_pred,y_test)

```

```
tol = [ 0.01,0.001,0.0001]
max_iter = [100,150,200]
param_grid = dict(tol = tol,max_iter=max_iter)
from sklearn.model_selection import GridSearchCV
grid_model = GridSearchCV(estimator=lre,cv=5,param_grid = param_grid)
x = preprocessing.MinMaxScaler().fit_transform(x)
gridsearch_result = grid_model.fit(x,y)
best_score,best_param = grid_model.best_score_,grid_model.best_params_
best_score,best_param
```
